```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
import os # Import the os module for directory operations

# ==========================================================
# CONFIG
# ==========================================================
SYMBOL         = "^NSEI"
TIMEFRAME      = "1h"

INITIAL_CAPITAL = 1_000_000   # ₹10L notional
RISK_PER_TRADE  = 0.10

BB_LENGTH = 20
BB_MULT   = 2.0

# ----------------------------------------------------------
# CHARGES CONFIG  (all rates are per-rupee of turnover
#                  unless marked as flat ₹ per order)
# ----------------------------------------------------------
# Instrument assumed: NSE Index Futures (NIFTY)
# Source: Zerodha / NSE brokerage calculator (2024)

BROKERAGE_FLAT    = 20.0        # ₹20 flat per executed order (like Zerodha)
# Alternatively use percentage: set BROKERAGE_PCT = 0.0003 and BROKERAGE_FLAT = None

STT_RATE          = 0.000125    # 0.0125% on SELL-side turnover (Futures)
EXCHANGE_CHARGE   = 0.000019    # 0.0019% both sides (NSE Futures)
SEBI_CHARGE       = 0.000001    # ₹10 per crore = 0.000001 on both sides
STAMP_DUTY_RATE   = 0.00002     # 0.002% on BUY-side turnover (Futures/Intraday)
GST_RATE          = 0.18        # 18% GST applies on: Brokerage + Exchange charges + SEBI charges

# ----------------------------------------------------------

def compute_charges(price: float, qty: float, side: str) -> dict:
    """
    Compute all applicable charges for one order.

    Parameters
    ----------
    price : float  – execution price
    qty   : float  – number of units (can be fractional for simulation)
    side  : str    – 'BUY' or 'SELL'

    Returns
    -------
    dict with individual charge components and 'total'
    """
    turnover = abs(price * qty)

    # 1. Brokerage
    brokerage = BROKERAGE_FLAT  # flat per order

    # 2. STT — only on SELL side for Futures
    stt = turnover * STT_RATE if side == "SELL" else 0.0

    # 3. Exchange Transaction Charges — both sides
    exchange = turnover * EXCHANGE_CHARGE

    # 4. SEBI Regulatory Charges — both sides
    sebi = turnover * SEBI_CHARGE

    # 5. Stamp Duty — only on BUY side
    stamp = turnover * STAMP_DUTY_RATE if side == "BUY" else 0.0

    # 6. GST — 18% on (Brokerage + Exchange + SEBI)
    gst = (brokerage + exchange + sebi) * GST_RATE

    total = brokerage + stt + exchange + sebi + stamp + gst

    return {
        "brokerage" : brokerage,
        "stt"       : stt,
        "exchange"  : exchange,
        "sebi"      : sebi,
        "stamp_duty": stamp,
        "gst"       : gst,
        "total"     : total,
    }

# ==========================================================
# FETCH DATA (Yahoo Finance – last 720 days)
# ==========================================================
end   = dt.datetime.now()
start = end - dt.timedelta(days=720)

print("Fetching NIFTY 50 (Yahoo Finance, 1H)...")

df = yf.download(
    SYMBOL,
    start=start,
    end=end,
    interval=TIMEFRAME,
    auto_adjust=True,
    progress=False
)

if df.empty:
    raise ValueError("No data returned from Yahoo Finance")

# Flatten MultiIndex columns
if isinstance(df.columns, pd.MultiIndex):
    df.columns = df.columns.get_level_values(0)

df.rename(columns={
    "Open": "open", "High": "high",
    "Low": "low",  "Close": "close", "Volume": "volume"
}, inplace=True)

df = df[["open", "high", "low", "close", "volume"]]

# ==========================================================
# INDICATORS — Bollinger Bands
# ==========================================================
df["basis"] = df["close"].rolling(BB_LENGTH).mean()
df["stdev"] = df["close"].rolling(BB_LENGTH).std()
df["upper"] = df["basis"] + BB_MULT * df["stdev"]
df["lower"] = df["basis"] - BB_MULT * df["stdev"]
df.dropna(inplace=True)

# ==========================================================
# SIGNALS
# ==========================================================
df["long_signal"] = (
    (df["close"] > df["upper"]) |
    ((df["low"] < df["lower"]) & (df["close"] > df["lower"]))
)
df["short_signal"] = (
    (df["close"] < df["lower"]) |
    ((df["high"] > df["upper"]) & (df["close"] < df["upper"]))
)

# ==========================================================
# BACKTEST ENGINE
# ==========================================================
capital        = INITIAL_CAPITAL
side           = None
entry_price    = 0
position_size  = 0

equity_curve   = []
trade_pnls     = []

# Charge accumulator for reporting
total_charges_paid = {
    "brokerage" : 0.0,
    "stt"       : 0.0,
    "exchange"  : 0.0,
    "sebi"      : 0.0,
    "stamp_duty": 0.0,
    "gst"       : 0.0,
    "total"     : 0.0,
}

def add_charges(charges: dict):
    for k in total_charges_paid:
        total_charges_paid[k] += charges[k]

for i in range(len(df)):
    row = df.iloc[i]
    o, h, l, c = row["open"], row["high"], row["low"], row["close"]

    # -------------------- EXIT --------------------
    if side:
        exit_price = None

        # Mean-reversion exit
        if (side == "LONG"  and c <= row["basis"]) or \
           (side == "SHORT" and c >= row["basis"]):
            exit_price = o

        # Stop-loss (2%)
        if (side == "LONG"  and l <= entry_price * 0.98) or \
           (side == "SHORT" and h >= entry_price * 1.02):
            exit_price = o

        if exit_price is not None:
            gross_pnl = (exit_price - entry_price) * position_size * (1 if side == "LONG" else -1)

            # Charges on exit order
            exit_side    = "SELL" if side == "LONG" else "BUY"
            exit_charges = compute_charges(exit_price, position_size, exit_side)
            add_charges(exit_charges)

            net_pnl = gross_pnl - exit_charges["total"]
            capital += net_pnl

            trade_pnls.append(net_pnl)

            side          = None
            position_size = 0

    # -------------------- ENTRY --------------------
    if not side:
        if   df["long_signal"].iloc[i]:  side = "LONG"
        elif df["short_signal"].iloc[i]: side = "SHORT"

        if side:
            entry_price    = o
            risk_amount    = capital * RISK_PER_TRADE
            stop_distance  = entry_price * 0.02
            position_size  = risk_amount / stop_distance

            # Charges on entry order
            entry_side     = "BUY" if side == "LONG" else "SELL"
            entry_charges  = compute_charges(entry_price, position_size, entry_side)
            add_charges(entry_charges)

            capital -= entry_charges["total"]

    # -------------------- EQUITY --------------------
    unrealized = (
        (c - entry_price) * position_size * (1 if side == "LONG" else -1)
        if side else 0
    )
    equity_curve.append(capital + unrealized)

# ==========================================================
# PERFORMANCE METRICS
# ==========================================================
equity   = pd.Series(equity_curve, index=df.index)
returns  = equity.pct_change().dropna()

total_return = (equity.iloc[-1] / INITIAL_CAPITAL - 1) * 100
max_drawdown = (equity / equity.cummax() - 1).min() * 100

wins   = [p for p in trade_pnls if p > 0]
losses = [p for p in trade_pnls if p < 0]

win_rate      = (len(wins) / len(trade_pnls)) * 100 if trade_pnls else 0
avg_win       = np.mean(wins)   if wins   else 0
avg_loss      = np.mean(losses) if losses else 0
profit_factor = sum(wins) / abs(sum(losses)) if losses else np.inf

sharpe = (
    (returns.mean() / returns.std()) * np.sqrt(252 * 6.5)
    if returns.std() != 0 else 0
)

# ==========================================================
# PRINT RESULTS
# ==========================================================
print("\n" + "═"*65)
print(" NIFTY 50 STRATEGY PERFORMANCE  (Yahoo Finance · 1H)")
print("═"*65)
print(f"  Initial Capital       : ₹{INITIAL_CAPITAL:>14,.0f}")
print(f"  Final Capital         : ₹{equity.iloc[-1]:>14,.0f}")
print(f"  Total Return          : {total_return:>13.2f}%")
print("-"*65)
print(f"  Total Trades          : {len(trade_pnls):>14}")
print(f"  Win Rate              : {win_rate:>13.2f}%")
print(f"  Average Win           : ₹{avg_win:>14,.0f}")
print(f"  Average Loss          : ₹{avg_loss:>14,.0f}")
print(f"  Profit Factor         : {profit_factor:>14.2f}")
print("-"*65)
print(f"  Sharpe Ratio          : {sharpe:>14.2f}")
print(f"  Max Drawdown          : {max_drawdown:>13.2f}%")
print("═"*65)
print(" CHARGES BREAKDOWN (Total Paid Across All Trades)")
print("═"*65)
print(f"  Brokerage (₹20/order) : ₹{total_charges_paid['brokerage']:>14,.2f}")
print(f"  STT (0.0125% sell)    : ₹{total_charges_paid['stt']:>14,.2f}")
print(f"  Exchange Charges      : ₹{total_charges_paid['exchange']:>14,.2f}")
print(f"  SEBI Charges          : ₹{total_charges_paid['sebi']:>14,.2f}")
print(f"  Stamp Duty (0.002%)   : ₹{total_charges_paid['stamp_duty']:>14,.2f}")
print(f"  GST (18%)             : ₹{total_charges_paid['gst']:>14,.2f}")
print("-"*65)
print(f"  TOTAL CHARGES         : ₹{total_charges_paid['total']:>14,.2f}")
pct_of_capital = total_charges_paid['total'] / INITIAL_CAPITAL * 100
print(f"  As % of Initial Cap   : {pct_of_capital:>13.2f}%")
print("═"*65)

# ==========================================================
# PLOTS
# ==========================================================
drawdown = (equity / equity.cummax() - 1) * 100

fig, (ax1, ax2) = plt.subplots(
    2, 1,
    figsize=(16, 9),
    sharex=True,
    gridspec_kw={"height_ratios": [3, 1], "hspace": 0.06}
)

fig.patch.set_facecolor("#0d1117")
for ax in (ax1, ax2):
    ax.set_facecolor("#0d1117")
    ax.tick_params(colors="#c9d1d9", labelsize=9)
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    for spine in ("left", "bottom"):
        ax.spines[spine].set_color("#30363d")

# ── Equity Curve ──────────────────────────────────────────
ax1.plot(equity.index, equity.values, lw=1.6, color="#58a6ff", zorder=3)
ax1.fill_between(equity.index, INITIAL_CAPITAL, equity.values,
                 where=(equity.values >= INITIAL_CAPITAL),
                 interpolate=True, color="#238636", alpha=0.25, zorder=2)
ax1.fill_between(equity.index, INITIAL_CAPITAL, equity.values,
                 where=(equity.values < INITIAL_CAPITAL),
                 interpolate=True, color="#da3633", alpha=0.25, zorder=2)

ax1.axhline(INITIAL_CAPITAL, color="#8b949e", lw=0.8, ls="--", label="Initial Capital")

# Annotate final value
final_val = equity.iloc[-1]
color_final = "#3fb950" if final_val >= INITIAL_CAPITAL else "#f85149"
ax1.annotate(
    f"₹{final_val:,.0f}  ({total_return:+.2f}%)",
    xy=(equity.index[-1], final_val),
    xytext=(-120, 18), textcoords="offset points",
    fontsize=9, color=color_final, fontweight="bold",
    arrowprops=dict(arrowstyle="->", color=color_final, lw=1),
)

ax1.set_ylabel("Portfolio Value  (₹)", color="#c9d1d9", fontsize=10)
ax1.yaxis.set_major_formatter(
    plt.FuncFormatter(lambda x, _: f"₹{x/1e5:.1f}L")
)
ax1.grid(axis="y", color="#21262d", lw=0.6, zorder=0)
ax1.grid(axis="x", color="#21262d", lw=0.4, zorder=0)
ax1.legend(fontsize=8, facecolor="#161b22", edgecolor="#30363d", labelcolor="#8b949e")

ax1.set_title(
    "NIFTY 50  ·  Bollinger Band Strategy  ·  1H  ·  Full Taxes & Charges Applied",
    color="#e6edf3", fontsize=12, fontweight="bold", pad=12
)

# Stats box
stats_txt = (
    f"Trades: {len(trade_pnls)}    Win Rate: {win_rate:.1f}%\n"
    f"Sharpe: {sharpe:.2f}    PF: {profit_factor:.2f}\n"
    f"Total Charges: ₹{total_charges_paid['total']:,.0f}"
)
ax1.text(0.01, 0.97, stats_txt,
         transform=ax1.transAxes, fontsize=8,
         verticalalignment="top",
         color="#8b949e",
         bbox=dict(boxstyle="round,pad=0.4", facecolor="#161b22",
                   edgecolor="#30363d", alpha=0.9))

# ── Drawdown ──────────────────────────────────────────────
ax2.fill_between(drawdown.index, 0, drawdown.values,
                 color="#da3633", alpha=0.55, zorder=2)
ax2.plot(drawdown.index, drawdown.values, lw=0.8, color="#f85149", zorder=3)

ax2.axhline(0, color="#30363d", lw=0.8)
ax2.set_ylabel("Drawdown  (%)", color="#c9d1d9", fontsize=10)
ax2.set_ylim(drawdown.min() * 1.15, 1)
ax2.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"{x:.1f}%"))
ax2.grid(axis="y", color="#21262d", lw=0.6, zorder=0)
ax2.grid(axis="x", color="#21262d", lw=0.4, zorder=0)

# Annotate max drawdown
min_dd_idx = drawdown.idxmin()
ax2.annotate(
    f"Max DD\n{max_drawdown:.2f}%",
    xy=(min_dd_idx, drawdown.min()),
    xytext=(30, -18), textcoords="offset points",
    fontsize=7.5, color="#f85149", fontweight="bold",
    arrowprops=dict(arrowstyle="->", color="#f85149", lw=0.9),
)

ax2.set_xlabel("Date", color="#c9d1d9", fontsize=10)

# Create the directory if it doesn't exist
os.makedirs('/mnt/user-data/outputs/', exist_ok=True)

plt.savefig("/mnt/user-data/outputs/equity_curve.png",
            dpi=150, bbox_inches="tight", facecolor="#0d1117")
plt.show()
print("\nChart saved → equity_curve.png")
```
